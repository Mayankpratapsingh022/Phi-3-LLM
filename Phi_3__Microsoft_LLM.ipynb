```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
# Print the tokenizer's vocab size

print(f"Vocabulary size: {tokenizer.vocab_size}")
```

```python
# Test the tokenizer with some example words

words = ['hello','world','AI',"learning"]

for word in words:
  print(f"Tokenizer '{word}' : {tokenizer.encode(word)}")
```

```python
# Test the tokenizer witht a sentence


sentence = "Artificial Intelligence is transforming the world."
encoded_sentence = tokenizer.encode(sentence)
decoded_sentence = tokenizer.decode(encoded_sentence)

print(f"\nOriginal sentence: {sentence}")
print(f"Tokenized sentence: {encoded_sentence}")
print(f"Decoded back: {decoded_sentence}")
```

```python
# Encode the sentence to get the token ids

encoded_sentence = tokenizer.encode(sentence)

# Decode each token id back to its respective token

tokens = [tokenizer.decode([token_id]) for token_id in encoded_sentence]


# Create a mapping of token IDs to tokens

token_id_map = list(zip(encoded_sentence,tokens))

# Print the token IDs and their corresponding tokens

for token_id, token in token_id_map:
  print(f"Token ID: {token_id}, Token: '{token}")

```

```python
from transformers import pipeline

# Create a pipeline

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False

)
```

```python
# The prompt (user input/query)
messages = [
    {"role": "system", "content": "You are a stand-up comedian"},
    {"role": "user", "content": "Create a funny joke about chickens."},
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
